# Imshow

Display data as an image, i.e., on a 2D regular raster.

You need to use the clabel command. You need to take care of the `origin` of the image in the imshow command and use a colorbar.

Starting from the code below, try to reproduce the graphic on the right taking care of colormap, image interpolation and origin.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x,y): return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 10
x = np.linspace(-3,3,4*n)
y = np.linspace(-3,3,3*n)
X,Y = np.meshgrid(x,y)
plt.imshow(f(X,Y))
plt.show()
```

````python
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
````

````python
def f(x,y):
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 10
x = np.linspace(-3,3,8*n)
y = np.linspace(-3,3,6*n)
X,Y = np.meshgrid(x,y)
Z = f(X,Y)

plt.axes([0.025,0.025,0.95,0.95])
plt.imshow(Z,interpolation='nearest', cmap='bone', origin='lower')
plt.colorbar(shrink=.75)

plt.xticks([]), plt.yticks([])

plt.show()
````
