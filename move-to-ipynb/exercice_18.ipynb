# Pie Charts

You need to modify Z.

Starting from the code below, try to reproduce the graphic on the right taking care of colors and slices size.

```python
n = 20
Z = np.random.uniform(0,1,n)
plt.pie(Z)
plt.show()
```

````python
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
````

````python
n = 20
Z = np.ones(n)
Z[-1] *= 2

plt.axes([0.025, 0.025, 0.95, 0.95])

plt.pie(Z, explode=Z*.05, colors=['%f' % (i/float(n)) for i in range(n)],
        wedgeprops={"linewidth": 1, "edgecolor": "black"})
plt.gca().set_aspect('equal')
plt.xticks([]), plt.yticks([])

plt.show()
````

# some changes

````python
n = 42
Z = np.ones(n)

plt.axes([0.025, 0.025, 0.95, 0.95])

plt.pie(Z, explode=Z*.4, colors=['%f' % (i/float(n+10)) for i in range(n)],
        wedgeprops={"linewidth": .75, "edgecolor": "yellow"})
plt.gca().set_aspect('equal')
plt.xticks([]), plt.yticks([])

plt.show()
````

````python
n = 8
Z = np.ones(n)

plt.axes([0.025, 0.025, 0.95, 0.95])

cmap = plt.get_cmap("tab20c")
colors = cmap(np.arange(2)*3)

plt.pie(Z, explode=Z*.4, colors=colors,
        wedgeprops={"linewidth": 10, "edgecolor": "#FFD700"})
plt.gca().set_aspect('equal')
plt.xticks([]), plt.yticks([])

plt.show()
````
